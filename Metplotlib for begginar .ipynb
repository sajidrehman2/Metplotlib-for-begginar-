# Figure size and DPI
  - Figure size ( figsize ) determines the size of the figure in inches. This gives the amount of space the axes (and other elements) have inside the figure. The default figure size is (6.4, 4.8) inches in matplotlib 2. A larger figure size will allow for longer texts, more axes or more ticklabels to be shown.

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 3))  # Create a figure with a specified size
axes = fig.add_axes([0, 0, 1, 1])  # Add axes to the figure
axes.plot(x, y)  # Plot the data on the axes
plt.show()  # Display the figure
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
num_points = 25
x = np.random.rand(num_points)
y = np.random.rand(num_points)

# Define colors and markers
colors = ['green', 'red', 'yellow', 'blue']
markers = ['o', 's', 'D', '^']

# Create scatter plot with different colors and markers
for i, (color, marker) in enumerate(zip(colors, markers)):
    plt.scatter(x[i::4], y[i::4], color=color, marker=marker, label=f'{color} {marker}')

# Add legend
plt.legend()

# Add labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Scatter Plot with Different Colors and Markers')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data generation
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a figure and a 3D axis
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot a 3D surface with customizations
surf = ax.plot_surface(x, y, z, cmap='plasma', edgecolor='none', alpha=0.9)
ax.set_xlabel('X Axis', fontsize=12, labelpad=10)
ax.set_ylabel('Y Axis', fontsize=12, labelpad=10)
ax.set_zlabel('Z Axis', fontsize=12, labelpad=10)
ax.set_title('Beautiful 3D Surface Plot', fontsize=15, pad=20)
ax.view_init(elev=30, azim=45)  # Adjust the view angle

# Add color bar which maps values to colors
cbar = fig.colorbar(surf, shrink=0.5, aspect=5)
cbar.set_label('Color Scale', fontsize=12)

# Additional customization for aesthetics
ax.grid(False)  # Turn off the grid for a cleaner look
ax.set_facecolor('white')  # Set background color

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Generate data for a simple human-like figure
# Head
head_x = [0]
head_y = [0]
head_z = [3]

# Body
body_x = [0, 0]
body_y = [0, 0]
body_z = [0, 2]

# Arms
arms_x = [-1, 1]
arms_y = [0, 0]
arms_z = [1, 1]

# Legs
legs_x = [-0.5, 0.5]
legs_y = [0, 0]
legs_z = [-1, -1]

# Combine all parts
x = head_x + body_x + arms_x + legs_x
y = head_y + body_y + arms_y + legs_y
z = head_z + body_z + arms_z + legs_z

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the points
ax.scatter(x, y, z, color='r', s=100)

# Set labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Human-like Figure')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Function to plot a sphere
def plot_sphere(ax, center, radius, color):
    u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
    x = center[0] + radius * np.cos(u) * np.sin(v)
    y = center[1] + radius * np.sin(u) * np.sin(v)
    z = center[2] + radius * np.cos(v)
    ax.plot_surface(x, y, z, color=color)

# Function to plot a cylinder
def plot_cylinder(ax, start, end, radius, color):
    v = np.linspace(0, 1, 100)
    theta = np.linspace(0, 2*np.pi, 100)
    v, theta = np.meshgrid(v, theta)
    x = start[0] + v * (end[0] - start[0]) + radius * np.cos(theta)
    y = start[1] + v * (end[1] - start[1]) + radius * np.sin(theta)
    z = start[2] + v * (end[2] - start[2])
    ax.plot_surface(x, y, z, color=color)

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot head (sphere)
plot_sphere(ax, center=[0, 0, 1.5], radius=0.2, color='blue')

# Plot body (cylinder)
plot_cylinder(ax, start=[0, 0, 0], end=[0, 0, 1.2], radius=0.1, color='red')

# Plot arms (cylinders)
plot_cylinder(ax, start=[-0.3, 0, 1.1], end=[0, 0, 0.6], radius=0.05, color='green')
plot_cylinder(ax, start=[0.3, 0, 1.1], end=[0, 0, 0.6], radius=0.05, color='green')

# Plot legs (cylinders)
plot_cylinder(ax, start=[-0.1, 0, 0], end=[0, 0, 0.6], radius=0.1, color='yellow')
plot_cylinder(ax, start=[0.1, 0, 0], end=[0, 0, 0.6], radius=0.1, color='yellow')

# Set axis labels
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Human Body Graph')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data generation using parametric equations for a bird's body-like shape
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
u, v = np.meshgrid(u, v)

# Parametric equations for the bird's body
x = 10 * np.sin(u) * np.sin(v)
y = 10 * np.cos(v)
z = 10 * np.cos(u) * np.sin(v)

# Create a figure and a 3D axis
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot a 3D surface with customizations
surf = ax.plot_surface(x, y, z, cmap='coolwarm', edgecolor='none', alpha=0.9)
ax.set_xlabel('X Axis', fontsize=12, labelpad=10)
ax.set_ylabel('Y Axis', fontsize=12, labelpad=10)
ax.set_zlabel('Z Axis', fontsize=12, labelpad=10)
ax.set_title('3D Bird\'s Body Plot', fontsize=15, pad=20)
ax.view_init(elev=20, azim=120)  # Adjust the view angle

# Add color bar which maps values to colors
cbar = fig.colorbar(surf, shrink=0.5, aspect=5)
cbar.set_label('Color Scale', fontsize=12)

# Additional customization for aesthetics
ax.grid(False)  # Turn off the grid for a cleaner look
ax.set_facecolor('white')  # Set background color

# Save the plot as a PNG file
fig.savefig('birds_body.png')

# Show plot
plt.show()
```
